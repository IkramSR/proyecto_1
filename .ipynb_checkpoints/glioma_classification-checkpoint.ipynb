```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/IKRAM/Documents/TCGA_InfoWithGrade.csv')
df.head(10)
```

# CreaciÃ³n del Ã¡rbol

No es necesario hacer limpieza ni preprocesamiento dado que no existen datos faltantes y las variables categÃ³ricas ya se encuentran codificadas; se define entonces la variable objetivo y las variables incluidas en el conjunto de entrenamiento.

Se crea el Ã¡rbol con los parÃ¡metros predeterminados y evaluamos la precisiÃ³n primero sin validaciÃ³n cruzada y despuÃ©s haciendo uso de esta. De los resultados mostrados se destaca que:
El modelo sin validaciÃ³n cruzada aparentemente da 100% de predicciones correctas, esto es practicamente sobreajuste ya que en la prÃ¡ctica es poco realista, este hecho se comprueba al aplicar validaciÃ³n cruzada y ver que la exactitud disminuye casi un 20%

```python
features = list(df.columns[1:])
y = df['Grade']
X = df[features]
model = DecisionTreeClassifier(random_state=0)
model.fit(X,y)
print(model.score(X,y))
print(cross_val_score(model,X,y,cv=10).mean())
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.30, random_state=0)
model1 = DecisionTreeClassifier(random_state=0)
model1.fit(x_train,y_train)
print(model1.score(x_test,y_test))
```

## Matriz de confusiÃ³n

Es importante hacer incapiÃ© en que es importante mantener sÃ­ una exactitud global lo mÃ¡s alta que se pueda pero especialmente debe tenerse como objetivo mantener los 'falsos negativos' al mÃ­nimo dado que potencialmente serÃ­a mÃ¡s riesgoso clasificar un glioblastoma multiforme como uno de bajo grado.

```python
from sklearn import metrics

def matriz_confusion(modelo):
    pred = modelo.predict(x_test)
    confusion_matrix = metrics.confusion_matrix(y_test, pred)
    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, 
                                                display_labels = ['LGG', 'GBM'])
    cm_display.plot()
    
matriz_confusion(model1)
```

## MÃ©tricas del modelo

Como se observa en el reporte de clasificaciÃ³n, para la primer clase la precision es mayor que el recall, lo que significa que el Ã¡rbol no detecta la clase muy bien, pero cuando lo hace es altamente confiable. Para la segunda clase sucede al contrario, una precisiÃ³n menor al recall significa que detecta bien dicha clase, pero tambiÃ©n incluye muestras de la primer clase.

```python
from sklearn.metrics import recall_score, precision_score

pred = model1.predict(x_test)
print(f"Sensibilidad del modelo: {round(recall_score(y_test, pred)*100,2)}%") 
print(f"PrecisiÃ³n del modelo: {round(precision_score(y_test, pred)*100,2)}%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred, target_names=['Lower-Grade Glioma (LGG)','Glioblastoma Multiforme (GBM)']))
```

## Importancia de caracterÃ­sticas y visualizaciÃ³n del Ã¡rbol

La visualizaciÃ³n del Ã¡rbol resulta difÃ­cil al no establecer restricciones en los parÃ¡metros del modelo, por ejemplo, max_depth. Aparentemente las variables correspondientes a los genes FUBP1, RB1 y FAT4 no tienen influencia significativa sobre la clasificaciÃ³n.

```python
feature_importance = pd.DataFrame(model1.feature_importances_, index=features, columns=['Importance']).sort_values(by='Importance', ascending=False)
feature_importance
```

```python
feature_importance.plot.bar()
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(80,20))
_=tree.plot_tree(model1,
         feature_names=features,
         class_names={0:'Lower-Grade Glioma (LGG)',1:'Glioblastoma Multiforme (GBM)'},
         filled=True,
         impurity=True)
```

## OptimizaciÃ³n de hiperparÃ¡metros

```python
parameters = {'criterion':['gini','entropy'],
             'max_depth':range(3,7),
             'max_features':range(5,15),
             }
```

```python
model_grid=GridSearchCV(model1,
                        param_grid=parameters,
                        verbose=1,
                        cv=10)
model_grid.fit(x_train,y_train)
```

```python
model_grid.best_params_
```

```python
model_grid.best_score_
```

## MÃ©tricas del modelo con los nuevos parÃ¡metros

Como puede observarse todas las mÃ©tricas muestran mejores resultados y es posible visualizar el Ã¡rbol mÃ¡s claramente

```python
matriz_confusion(model_grid)
```

```python
pred_grid=model_grid.predict(x_test)
print(classification_report(y_test, pred_grid, target_names=['Lower-Grade Glioma (LGG)','Glioblastoma Multiforme (GBM)']))
```

```python
model_grid1 = DecisionTreeClassifier(max_depth= 4, max_features=10, random_state=0)
model_grid1.fit(x_train,y_train)
fig=plt.figure(figsize=(80,20))
_=tree.plot_tree(model_grid1,
         feature_names=features,
         class_names={0:'Lower-Grade Glioma (LGG)',1:'Glioblastoma Multiforme (GBM)'},
         filled=True,
         impurity=True)
```

## Random Forest

Generamos un bosque de decisiÃ³n y optimizamos hiperparÃ¡metros haciendo uso de RandomizedSearchCV

```python
from sklearn.ensemble import RandomForestClassifier
forest_glioma = RandomForestClassifier()
forest_glioma.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
pred_fg = forest_glioma.predict(x_test)
accuracy_score(y_test, pred_fg)
```

```python
from sklearn.model_selection import RandomizedSearchCV
random_grid = {'n_estimators': [50,100,150,200],
               'max_features': range(5,10),
               'max_depth': range(2,7),
               'bootstrap': [True, False]
              }
glioma_randomgrid = RandomizedSearchCV(estimator = forest_glioma, 
                                    param_distributions = random_grid, 
                                    n_iter = 20, cv = 10, verbose=2, 
                                    random_state=0)
```

```python
glioma_randomgrid.fit(x_train, y_train)
```

```python
glioma_randomgrid.best_params_
```

```python
pred_gliomagrid=glioma_randomgrid.predict(x_test)
accuracy_score(y_test,pred_gliomagrid)
```

```python
print(classification_report(y_test, pred_gliomagrid, target_names=['Lower-Grade Glioma (LGG)','Glioblastoma Multiforme (GBM)']))
```

```python
matriz_confusion(glioma_randomgrid)
```

```python
forest_glioma_grid = RandomForestClassifier(n_estimators=100, 
                                       max_features=8, 
                                       max_depth=4, 
                                       bootstrap=False)
forest_glioma_grid.fit(x_train, y_train)
feature_importances = pd.DataFrame(forest_glioma_grid.feature_importances_, index=features, columns=['Importance']).sort_values(by='Importance', ascending=False)
feature_importances.plot.bar()
```

## Conclusiones

1. Es fÃ¡cil sobreajustar un Ã¡rbol de decisiÃ³n, se deben restringen ciertos parÃ¡metros y evaluar con validaciÃ³n cruzada.

2. El remuestreo aleatorio en el que se basan los bosques de decisiÃ³n hace que sea mÃ¡s dificil caer en el sobreajuste pero es importante elegir hiperparÃ¡metros adecuados para lograr una mejor precisiÃ³n.

3. Aunque tanto la precisiÃ³n global como individual (para cada clase) es practicamente la misma para ambos modelos (DecisionTree y RandomForest) la importancia de variables no es igual, por lo tanto el modelo a usar se deberÃ¡ seleccionar tomando en cuenta el grupo de variables mÃ¡s significativas que representen un menor costo de obtenciÃ³n; por ejemplo, serÃ­a mejor usar el modelo que da mayor importancia a la raza que al resultado de un estudio gÃ©netico especÃ­fico, siendo que ambos modelos tienen la misma precisiÃ³n.
